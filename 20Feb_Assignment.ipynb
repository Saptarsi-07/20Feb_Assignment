```python
## Q5)
from flask import Flask,redirect,url_for


app = Flask(__name__)

@app.route("/under/<int:n>")
def test1(n):
    return "Age:{} UNDERAGE FOR VOTING".format(n)

@app.route("/above/<int:n>")
def test2(n):
    return "Age:{} ELIGIBLE TO VOTE".format(n)



##to check age
@app.route("/age/<int:n>")
def results(n):
    outcome=""
    if(n<18):
        outcome="test1"
    else:
        outcome="test2"
    
    return redirect(url_for(outcome,n=n))


    
    

if __name__=="__main__":
    app.run(host="0.0.0.0")

```
